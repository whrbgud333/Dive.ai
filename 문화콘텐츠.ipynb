## 3. JSON 스키마 상세

### 3-1. 원천데이터 (TS_\* / VS_\*)

스크립트 원문 텍스트를 문장 단위 배열로 저장합니다.

```json
{
  "id":    "01_0017",        // 파일명과 동일한 고유 ID
  "type":  "movie",          // 콘텐츠 유형: movie | series | novel | comic
  "title": "아이들이 보인다",  // 작품 제목
  "year":  2003,              // 제작/출판 연도
  "text": [                   // 스크립트/본문 문장 배열 (캐릭터명은 C001 등 익명 처리)
    "C001은 지하철 노약자석 한구석에서 졸고 있다.",
    "C001은 졸다가 눈을 가늘게 떴을 때 어떤 엄마와 아이를 본다.",
    ...
  ]
}
```

### 3-2. 라벨링데이터 (TL_\* / VL_\*)

원천데이터에 스토리 분석 레이블과 씬(unit) 단위 구조를 추가합니다.

```json
{
  "id":            "01_0017",
  "type":          "movie",
  "title":         "아이들이 보인다",
  "genre":         ["스릴러", "공포(호러)", "드라마"],  // 복수 장르
  "year":          2003,
  "theme":         "발견",                             // 작품 핵심 테마
  "concept":       "C001은 퇴근길에...",               // 한 줄 줄거리
  "structure":     "스토리헬퍼 15단계",                // 스토리 구조 프레임워크
  "motif":         "갑작스런 사고",                    // 주요 모티프
  "main_character":"방어적임",                         // 주인공 성격 유형
  "conflict":      "C001은 집에 보이는 귀신이...",    // 주요 갈등
  "characters":    ["C001", "C002", ...],              // 등장인물 목록
  "units": [
    {
      "id":         "01_0017_01",     // 씬 ID
      "next_id":    "01_0017_02",     // 다음 씬 ID (연결 구조)
      "characters": ["C001", "C003"],// 해당 씬 등장인물
      "stage":      "Opening Salvo", // 스토리 단계 (영문)
      "storyline":  "C001은 지하철 안에서...", // 씬 요약
      "unit_motif": "긴장감",          // 씬 모티프
      "story_scripts": [
        {
          "type":      "narrative",   // narrative | dialogue
          "location":  "지하철 열차 안",
          "character": ["C001"],
          "act":       "공개하다/드러내다", // 행동 레이블
          "emotion":   "피곤하다",          // 감정 레이블
          "content":   "C001은 지하철 노약자석 한구석에서 졸고 있다."
        },
        ...
      ]
    },
    ...
  ]
}
```

## 4. 필드 정리

> 실제 데이터 확인 결과, **영화/시리즈/소설/만화 4개 카테고리 모두 동일한 스키마**를 공유합니다.

### 4-1. 원천데이터 필드

| 필드 | 타입 | 설명 |
|------|------|------|
| `id` | string | 고유 ID. `{카테고리번호}_{작품번호}` 형식 (예: `01_0017`). 라벨링데이터와 1:1 매핑 |
| `type` | string | 콘텐츠 유형: `movie` / `series` / `novel` / `comic` |
| `title` | string | 작품명 |
| `year` | int | 제작/출판 연도 |
| `text` | array\<string\> | 전체 스크립트/본문을 문장 단위 배열로 저장. 캐릭터명은 C001 등으로 익명 처리 |

### 4-2. 라벨링데이터 필드 — 작품 레벨

| 필드 | 타입 | 설명 |
|------|------|------|
| `id` | string | 원천데이터와 동일한 고유 ID |
| `type` | string | 콘텐츠 유형: `movie` / `series` / `novel` / `comic` |
| `title` | string | 작품명 |
| `year` | int | 제작/출판 연도 |
| `genre` | array\<string\> | 복수 장르 배열 (예: `["스릴러", "공포(호러)", "드라마"]`) |
| `theme` | string | 작품 핵심 테마 키워드 (예: `"발견"`) |
| `concept` | string | 한 줄 줄거리 요약 |
| `structure` | string | 스토리 구조 프레임워크 (예: `"스토리헬퍼 15단계"`) |
| `motif` | string | 작품 전체 모티프 (예: `"갑작스런 사고"`) |
| `main_character` | string | 주인공 성격 유형 (예: `"방어적임"`) |
| `conflict` | string | 주요 갈등 서술 |
| `characters` | array\<string\> | 등장인물 ID 목록 (예: `["C001", "C002", ...]`) |
| `units` | array\<object\> | 씬(unit) 단위 분석 배열 → 아래 4-3 참고 |

### 4-3. 라벨링데이터 필드 — 씬(unit) 레벨

| 필드 | 타입 | 설명 |
|------|------|------|
| `units[].id` | string | 씬 고유 ID (예: `"01_0017_01"`) |
| `units[].prev_id` | string | 이전 씬 ID (첫 씬은 null) |
| `units[].next_id` | string | 다음 씬 ID (마지막 씬은 null) |
| `units[].characters` | array\<string\> | 해당 씬 등장인물 ID 목록 |
| `units[].stage` | string | 스토리 단계명, 영문 (스토리헬퍼 15단계 기준, 예: `"Opening Salvo"`) |
| `units[].storyline` | string | 씬 내용 요약 |
| `units[].unit_motif` | string | 씬 단위 모티프 (예: `"긴장감"`) |
| `units[].causality` | string | 씬 간 인과관계 서술 |
| `units[].story_scripts` | array\<object\> | 문장별 상세 레이블 배열 → 아래 4-4 참고 |

### 4-4. 라벨링데이터 필드 — 스크립트(story_scripts) 레벨

| 필드 | 타입 | 설명 |
|------|------|------|
| `units[].story_scripts[].type` | string | 문장 유형: `narrative` (지문) / `dialogue` (대사) |
| `units[].story_scripts[].location` | string | 해당 문장의 배경 장소 (예: `"지하철 열차 안"`) |
| `units[].story_scripts[].character` | array\<string\> | 해당 문장의 등장인물 ID 목록 |
| `units[].story_scripts[].act` | string | 행동 레이블 (예: `"공개하다/드러내다"`) |
| `units[].story_scripts[].emotion` | string | 감정 레이블 (예: `"피곤하다"`) |
| `units[].story_scripts[].content` | string | 원문 문장 텍스트 |

## 5. 카테고리별 작품 수

파일명 패턴 `{카테고리번호}_{작품번호}.json` 에서 고유 작품을 식별 (Training + Validation 중복 제거).

> 각 작품은 JSON 파일 1개로 저장되며, 파일 수(7,122개)와 차이가 나는 것은 Training/Validation 분할 때문입니다.

## 6. 데이터 로드 예시 코드

````python
import json
import os
from pathlib import Path

BASE_DIR = Path("문화콘텐츠 스토리 데이터/145.다양한_문화콘텐츠_스토리_데이터/01-1.정식개방데이터")

# 카테고리 매핑
CATEGORIES = {
    "01": "영화",
    "02": "시리즈",
    "03": "소설",
    "04": "만화",
}

def load_pair(split: str, category: str, file_id: str):
    """원천데이터와 라벨링데이터를 ID로 한 쌍 불러오기.
    
    split: 'Training' | 'Validation'
    category: '01'~'04'
    file_id: e.g. '01_0017'
    """
    prefix_src = "TS" if split == "Training" else "VS"
    prefix_lbl = "TL" if split == "Training" else "VL"
    cat_name = CATEGORIES[category]

    src_path = BASE_DIR / split / "01.원천데이터" / f"{prefix_src}_{category}.{cat_name}" / f"{file_id}.json"
    lbl_path = BASE_DIR / split / "02.라벨링데이터" / f"{prefix_lbl}_{category}.{cat_name}" / f"{file_id}.json"

    with open(src_path, encoding="utf-8") as f:
        source = json.load(f)
    with open(lbl_path, encoding="utf-8") as f:
        label = json.load(f)

    return source, label


def iter_all(split: str = "Training", category: str = "01"):
    """특정 split/카테고리의 모든 (원천, 라벨링) 쌍을 순회."""
    prefix_src = "TS" if split == "Training" else "VS"
    cat_name = CATEGORIES[category]
    src_dir = BASE_DIR / split / "01.원천데이터" / f"{prefix_src}_{category}.{cat_name}"

    for json_file in sorted(src_dir.glob("*.json")):
        file_id = json_file.stem
        yield load_pair(split, category, file_id)


# 사용 예시
src, lbl = load_pair("Training", "01", "01_0017")
print("제목:", src["title"])
print("장르:", lbl["genre"])
print("테마:", lbl["theme"])
print("씬 수:", len(lbl["units"]))
print("첫 번째 씬 stage:", lbl["units"][0]["stage"])
````

````python
# 전체 파일 수 확인
stats = {}
for split in ["Training", "Validation"]:
    for cat in CATEGORIES:
        prefix = "TS" if split == "Training" else "VS"
        cat_name = CATEGORIES[cat]
        d = BASE_DIR / split / "01.원천데이터" / f"{prefix}_{cat}.{cat_name}"
        count = len(list(d.glob("*.json"))) if d.exists() else 0
        stats[f"{split}/{cat_name}"] = count

for k, v in stats.items():
    print(f"{k}: {v}개")
print("합계:", sum(stats.values()), "(원천 기준)")
````
