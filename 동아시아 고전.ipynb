## 3. 파일 형식

### 3-1. 원천데이터 — `.txt` (단락 단위 고전 텍스트)

- 파일 하나 = 고전 작품의 **단락(문단) 1개**
- 한국어 번역문으로 저장
- 파일명 패턴: `ST01_{국가코드}_{작품코드}_{섹션코드}_{단락코드}.txt`

```
예) ST01_03_01_0041_04_0012.txt
     └─ ST01  : 데이터셋 코드
     └─ 03    : 국가 (01=한국, 02=중국, 03=일본)
     └─ 01    : 작품 분류
     └─ 0041  : 작품 번호
     └─ 04    : 섹션 코드
     └─ 0012  : 단락 번호
```

**내용 예시** (일본 고전):
```
그러고 나서야 구라우도가 있는 장소에 와서 "이제 내전으로 가겠습니다."
라고 말하고 수레를 타고 내전으로 향했다. 천황은 그를 부르시며 "경을 읽어주십시오"
라고 말씀하셨다. ...
```

### 3-2. 라벨링데이터 — `_info.json` (내용 분류 + 주제 정보)

- 원천 txt와 **동일한 기본 파일명** + `_info.json` 접미사
- 단락 텍스트에 대한 9가지 분류 레이블과 주제어·주제문 포함

```json
{
  "country": "03",         // 국가코드 (01=한국, 02=중국, 03=일본)
  "class_1": "01",         // 작품 분류
  "내용분류": {
    "감정":     ["해당없음"],          // 단락에서 표현된 감정
    "대화형태": ["설명"],              // 대화 유형: 명령/설명/답변/독백 등
    "행동":     ["반응"],              // 주요 행동 유형
    "인물관계": ["해당없음"],          // 등장인물 간 관계
    "캐릭터":   ["요괴", "왕족/귀족", "승려"],  // 등장 캐릭터 유형
    "직업/신분":["승려"],              // 인물 직업/신분
    "장르":     ["설화"],              // 작품 장르
    "사건/모티프":["기적", "종교"],    // 핵심 사건 또는 모티프
    "공간":     ["궁"]                 // 배경 공간
  },
  "주제어": ["구라우도", "내전", "천황", "법화경", "승강"],  // 핵심 키워드 목록
  "주제문": "에이지쓰가 내전으로 향하며 법화경을 읽자 천황의 병이 호전되어 ..."  // 단락 요약
}
```

## 7. Dive.ai에서의 활용 방법

> 동아시아 고전 데이터는 **서사 구조**가 아닌 **세계관·분위기 보강** 용도입니다.  
> 유저가 "조선", "중국 무협", "일본 설화" 등 **고전 장르를 명시적으로 선택했을 때만** 참조합니다.

### 7-1. 고전 장르 판별

```python
CLASSIC_GENRES = {"조선", "중국무협", "일본설화", "한국설화", "동아시아고전"}

def is_classic_genre(background: str) -> bool:
    return any(tag in background for tag in CLASSIC_GENRES)

# 예시
is_classic_genre("조선 시대")   # True  → 고전 데이터 참조
is_classic_genre("현대 스릴러") # False → 고전 데이터 미참조
```

### 7-2. 필드별 활용 방법

| 필드 | Dive.ai 활용 |
|------|-------------|
| `내용분류.장르` | 고전 RAG 필터링 키 (예: "설화", "무협"으로 검색 범위 한정) |
| `내용분류.사건/모티프` | RAG 유사 단락 검색 키 (예: "복수", "귀신"으로 매칭) |
| `내용분류.공간` | 로어북 배경 키워드 보강 + 이미지 생성 프롬프트 (예: "궁" → `royal palace, hanok architecture`) |
| `내용분류.캐릭터` + `직업/신분` | 로어북 인물 유형 보강 (예: "왕족/귀족", "승려", "요괴") |
| `내용분류.대화형태` | 고전 말투 few-shot 예시 (예: 명령체/존댓말 스타일 참고) |
| `주제문` | LLM 프롬프트에 세계관 힌트로 주입 (단락 요약 1문장) |
| `주제어` | 로어북 고유명사 키워드 (인물명, 지명 등 실명 포함) |

### 7-3. 프롬프트 주입 방식

고전 장르 선택 시, 문화콘텐츠 서사 구조 아래에 세계관 힌트를 **추가** 형태로 붙입니다.

```
[유저 입력] 조선 퇴마사 복수극 (장르: 스릴러, 배경: 조선)
```

## 7. 데이터 로드 예시 코드

````python
import json
from pathlib import Path

BASE_DIR = Path("동아시아 고전 데이터/70.동아시아_고전_스토리_데이터/3.개방데이터/1.데이터")

COUNTRY_CODE = {"01": "한국", "02": "중국", "03": "일본"}


def load_pair(split: str, country: str, group: str, file_stem: str):
    """원천(txt)과 라벨링(json)을 한 쌍 불러오기.

    split  : 'Training' | 'Validation'
    country: '한국' | '중국' | '일본'
    group  : '01' ~ '08' (국가별 그룹 번호)
    file_stem: 파일명 (확장자 제외), e.g. 'ST01_03_01_0041_04_0012'
    """
    prefix_src = "TS" if split == "Training" else "VS"
    prefix_lbl = "TL" if split == "Training" else "VL"

    src_dir = BASE_DIR / split / "01.원천데이터" / f"{prefix_src}_ST01_{country}_{group}_text"
    lbl_dir = BASE_DIR / split / "02.라벨링데이터" / f"{prefix_lbl}_ST01_{country}_{group}_info"

    src_text = (src_dir / f"{file_stem}.txt").read_text(encoding="utf-8")
    with open(lbl_dir / f"{file_stem}_info.json", encoding="utf-8") as f:
        label = json.load(f)

    return src_text, label


def iter_all(split: str = "Training", country: str = "한국", group: str = "01"):
    """특정 split/국가/그룹의 모든 (원천, 라벨링) 쌍을 순회."""
    prefix_src = "TS" if split == "Training" else "VS"
    src_dir = BASE_DIR / split / "01.원천데이터" / f"{prefix_src}_ST01_{country}_{group}_text"

    for txt_file in sorted(src_dir.glob("*.txt")):
        yield load_pair(split, country, group, txt_file.stem)


# 사용 예시
text, label = load_pair("Training", "일본", "01", "ST01_03_01_0041_04_0012")
print("원문:", text[:80], "...")
print("장르:", label["내용분류"]["장르"])
print("감정:", label["내용분류"]["감정"])
print("사건/모티프:", label["내용분류"]["사건/모티프"])
print("주제문:", label["주제문"])
````

````python
# 전체 파일 수 확인
stats = {}
for split in ["Training", "Validation"]:
    for country in ["한국", "중국", "일본"]:
        max_group = 7 if country == "한국" else 8
        prefix = "TS" if split == "Training" else "VS"
        total = 0
        for g in range(1, max_group + 1):
            d = BASE_DIR / split / "01.원천데이터" / f"{prefix}_ST01_{country}_{g:02d}_text"
            total += len(list(d.glob("*.txt"))) if d.exists() else 0
        stats[f"{split}/{country}"] = total

for k, v in stats.items():
    print(f"{k}: {v:,}개")
print("원천 합계:", f"{sum(stats.values()):,}개")
print("전체 합계 (원천+라벨링):", f"{sum(stats.values())*2:,}개")
````
